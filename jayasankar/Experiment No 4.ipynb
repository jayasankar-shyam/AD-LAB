# Program to predict the sepal width with sepal length

```python
import matplotlib.pyplot as plt
from scipy import stats
import pandas as pd
data=pd.read_csv('Iris.csv')
x=data.loc[:,'SepalLengthCm']
y=data.loc[:,'SepalWidthCm']
m,c,r,p,std_err=stats.linregress(x,y)
def y_value(x):
    return m*x+c
model=list(map(y_value,x))
plt.scatter(x,y,c='g',label="Datas")
plt.plot(x,model,c='r',label="Regression Line")
plt.title("Linear Regression(Iris Data)")
plt.xlabel("Sepal length in cm")
plt.ylabel("Sepal width in cm")
predicted=float(input("Enter the length of sepal : "))
print(f"The predicted sepal width is {y_value(predicted):.4f}")
plt.scatter(predicted,y_value(predicted),marker="o", color="blue",label="Predicted Value")
plt.legend(bbox_to_anchor = (1.4 , 1.03))
plt.show()
```
